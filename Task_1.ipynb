```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/dm_office_sales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# sns.set(style='white')
sns.displot(data=df,x='salary',bins=50,color='red',edgecolor='black',linewidth=0.9,ls='--')
```

```python
sns.displot(data=df,x='salary',kde=True)
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,errorbar='sd',hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```
